The **Mandelbrot set** ([/ˈmændəlbroʊt,  -brɒt/] is a two-dimensional set  with a relatively simple definition that exhibits great complexity, especially as it is magnified. It is popular for its aesthetic appeal and fractal structures. The set is defined in the complex plane as the complex numbers $c$ for which the function $f_{c}(z)=z^{2}+c$ does not diverge to infinity when iterated starting at $z=0$, i.e., for which the sequence $f_{c}(0)$, $f_{c}(f_{c}(0))$, etc., remains bounded in absolute value.

```python
# Importing necessary functions from the required module
from numpy import zeros, linspace, square, add, mod
from matplotlib.pyplot import xlabel, ylabel, figure, imshow, title, show
from numba import njit
```

```python
class MandelbrotSet:

    @njit
    def mandelbrot_set_function(c, max_iteration, z):

        for ITER in range(max_iteration):
            if 2 > abs(z) > -1.5:
                z = add(square(z),c)
            else:
                return ITER
        return 

    @staticmethod
    def image_generation(minimum_value_of_x, maximum_value_of_x, minimum_value_of_y, maximum_value_of_y, width_of_graph, height_of_graph):
        x = linspace(minimum_value_of_x, maximum_value_of_x, width_of_graph)
        y = linspace(minimum_value_of_y, maximum_value_of_y, height_of_graph)
        mset = zeros((height_of_graph, width_of_graph))

        for i in range(height_of_graph):
            for j in range(width_of_graph):
                mset[i,j] = MandelbrotSet.mandelbrot_set_function(complex(x[j], y[i]), 10, 0)
        
        return mset
```

```python
minimumValueOfX, maximumValueOfX, minimumValueOfY, maximumValueOfY,width,height = -2, 1,-1.5,1.5,100,100
```

```python
image = MandelbrotSet.image_generation(minimumValueOfX,maximumValueOfX, minimumValueOfY, maximumValueOfY, width, height)
```

```python
# @jit
def generate_image():
    figure(figsize=(16,16), dpi=700)
    imshow(image, extent=(minimumValueOfX,maximumValueOfX,minimumValueOfY,maximumValueOfY), cmap = 'inferno')
    title("Mandelbrot Set")
    xlabel("Real Axis")
    ylabel('Complex Axis')

    show() #Generating Image
```
